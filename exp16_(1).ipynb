```python

#1. Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
text = "Natural language processing is interesting"
print(word_tokenize(text))
```

```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize
text = "Python is easy to learn. Python is easy to understand. Python is easy to use."
print(sent_tokenize(text))
```

```python
#Stop Word Removal
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Steming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing","studies","running","gone","watched","coding"]
for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("playing"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("best"))
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("coding"))

##NEED TO EXPLORE FURTHER##
```

```python
#Part-Of-Speech (POS) Tagging

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words = word_tokenize("Python is a powerful programming language")
pos_tag1 = pos_tag(words)
print(pos_tag1)
```

```python
#Word Frequency Count
from nltk.probability import FreqDist
word = word_tokenize(text)
fdist = FreqDist(word)
print(fdist.most_common())
```
